## Plotting Functions

```python
# X values.
x = np.linspace(0.0, 100.0, 1001)

# Show.
x
```

```python
# Y values.
y = x**2

# Show.
y
```

```python
# Plot y vs x
plt.plot(x, y, label='$x^2$')

# Legend
plt.legend()

# Labels
plt.xlabel('x')
plt.ylabel('$x^2$')

# Title
plt.title('A Lovely Plot');
```

## Comparing Plots

```python
# x values
x = np.linspace(0.0, 100.0, 1001)

# y1 = x^2
y1 = x**2

# y2 = 2^x
y2 = 2**x
```

```python
# Plot x vs y1.
plt.plot(x, y1, label='$x^2$')

# Plot x vs y2.
plt.plot(x, y2, label='$2^x$')

# Legend
plt.legend()

# Labels
plt.xlabel('x')
plt.ylabel('y')

# Title
plt.title('Two Lovely Plots');
```

## Integers

```python
# x values
# https://numpy.org/doc/stable/reference/generated/numpy.arange.html
x = np.arange(1, 101, 1)

# Show.
x
```

```python
# The type.
x.dtype
```

```python
# x values
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
x = np.linspace(0, 10, 11, dtype=int)

# Show.
x
```

```python
# The type.
x.dtype
```

```python
# x cubed plus 2 x squared plus 1.
y = x**3 + 2 * x**2 + 1
```

```python
# Show y.
y
```

```python
# Type of y.
y.dtype
```

```python
# Plot y vs x
plt.plot(x, y, '.', label='$x^3 + 2x^2 + 1$')

# Legend
plt.legend()

# Labels
plt.xlabel('x')
plt.ylabel('y')

# Title
plt.title('Plotting ints.');
```
